```python
def mm(data, y_test):
    t0x = np.matrix(data).T
    t0x = np.reshape(t0x, (n,2))
    t0xx = y_test - t0x
    
    #t0xy = t0xx[:,0]-t0xx[:,1]
    t0xy = t0xx[:,0]-t0xx[:,1]
    return t0xy
```

```python
t0_x2 = mm(data_original['Predictions'].iloc[31], y_test)
t1_x2 = mm(data_original['Predictions'].iloc[30], y_test)
t2_x2 = mm(data_original['Predictions'].iloc[32], y_test)

t3_x2 = mm(data_original['Predictions'].iloc[39], y_test)
t4_x2 = mm(data_original['Predictions'].iloc[33], y_test)
```

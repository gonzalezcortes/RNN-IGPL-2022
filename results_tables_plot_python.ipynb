```python
results0 = data_original.loc[data_original.groupby(['Model'])['MSE'].idxmin()]
results0
```

```python
results = data_original.loc[data_original.groupby(['Model','t'])['MSE'].idxmin()]
results[['MSE','MAE']] = results[['MSE','MAE']].round(decimals=4)
results[['R2', 'Running Time']] = results[['R2','Running Time']].round(decimals=2)
results = results.sort_values(by=['t'])
results = results.drop(columns=['Configuration','Predictions'])
results
```

```python
results.to_latex(index=False)
```

```python
results_pivot = results.pivot(index='Model', columns='t', values=['MSE', 'Specifications','Running Time'])
results_pivot
```

## Best models -> t = 0

```python
#data_original = pd.read_csv(path+'results/final_results_25_01_2023-14_58_12.csv')
```

```python
#Best models
results_t0 = results[results['t']==0]
y_test_0 = np.genfromtxt(path+'results/y_test_array_0.csv', delimiter=',')
```

```python
tx0 = lambda n: np.squeeze(np.array(np.matrix(data_original['Predictions'].iloc[n]).T))
rt0 = np.matrix(list(map(tx0, results_t0.index)))
rt_df0 = pd.DataFrame(rt0.T, columns=[np.array(results_t0['Model'])])
rt_df0['real'] = y_test_0
rt_df0['t'] = np.array(list(range(1,len(rt_df0)+1)))
```

```python
rt_df0.to_csv(path+"/results/t_0.csv", index=False)
```
